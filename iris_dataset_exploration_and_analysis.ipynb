# 1: Loading & Understanding the Dataset


```python
# Importing the pandas library for data manipulation
import pandas as pd  

# Importing matplotlib and seaborn for data visualization
import seaborn as sns 
import matplotlib.pyplot as plt 

# Loading the iris dataset from seaborn library
iris = sns.load_dataset('iris') 

# Creating a pandas DataFrame called 'dataset' from the iris dataset
dataset = pd.DataFrame(iris)  
```

```python
# Displaying the first 10 rows of the dataset to examine the data structure
dataset.head(10)
```

# 2: Basic Statistical Descriptions

**Mean**

also known as the average, represents the central tendency of a dataset, calculated by summing all the data points and dividing by the total number of data points.

```python
# Printing the mean values of all of the numeric features
print(dataset.mean(numeric_only=True).to_string(header=False))
```

---

**Median**

often referred to as the middle number in a dataset, represents the value that separates the higher half from the lower half when all data points are arranged in ascending order.

```python
# Printing the median values of all of the numeric features
print(dataset.median(numeric_only=True).to_string(header=False))
```

---

**Mode**

also known as the most common number in the dataset or modal value, represents the value that occurs most frequently in the dataset. It indicates which value appears most frequently throughout the dataset, providing insight into the most prevalent value among the data points.

```python
# Printing the mode values of all of the numeric features
print(dataset.mode(numeric_only=True))
```

---

**Variance**

also known as the spread between numbers in the dataset, measures the extent to which each number deviates from the mean and from every other number in the dataset. It quantifies the dispersion or variability of the data points, providing insight into how much individual data points differ from the average value.

```python
# Printing the variance values of all of the numeric features
print(dataset.var(numeric_only=True).to_string(header=False))
```

---

**Standard Deviation**

often referred to as the dispersion of a dataset relative to its mean, is calculated as the square root of the variance. It provides a measure of how much individual data points deviate from the average value. A higher standard deviation indicates that the data points are further from the mean, resulting in greater variability or dispersion within the dataset. In other words, the more spread out the data, the higher the standard deviation.

```python
# Printing the standard deviation values of all of the numeric features
print(dataset.std(numeric_only=True).to_string(header=False))
```

# 3: Data Visualization

## **Histogram**


```python
# Creating the main plot
plt.figure(figsize=(10,10))

# Subplot histogram for Sepal Length
plt.subplot(2,2,1)
sns.histplot(dataset['sepal_length'], kde=True, color='blue')
plt.title('Sepal Length', fontsize=16)
plt.ylabel('Frequency')

# Subplot histogram for Sepal Width
plt.subplot(2,2,2)
sns.histplot(dataset['sepal_width'], kde=True, color='red')
plt.title('Sepal Width', fontsize=16)
plt.ylabel('Frequency')

# Subplot histogram for Petal Length
plt.subplot(2,2,3)
sns.histplot(dataset['petal_length'], kde=True, color='green')
plt.title('Petal Length', fontsize=16)
plt.ylabel('Frequency')

# Subplot histogram for Petal Width
plt.subplot(2,2,4)
sns.histplot(dataset['petal_width'], kde=True, color='orange')
plt.title('Petal Width', fontsize=16)
plt.ylabel('Frequency')

# Adjusting the plot gaps
plt.subplots_adjust(wspace=0.4, hspace=0.4)

# Showing the grids
plt.grid(True)

# Showing the main plot
plt.show()
```

---

## **Box Plot**

```python
# Creating the main plot
plt.figure(figsize=(12, 12))

# Subplot boxplot for Sepal Length
plt.subplot(2, 2, 1)
sns.boxplot(x='species', y='sepal_length', data=dataset, color='blue')
plt.title('Sepal Length', fontsize=16)
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.grid(True)

# Subplot boxplot for Sepal Width
plt.subplot(2, 2, 2)
sns.boxplot(x='species', y='sepal_width', data=dataset, color='red')
plt.title('Sepal Width', fontsize=16)
plt.xlabel('Species')
plt.ylabel('Sepal Width')
plt.grid(True)

# Subplot boxplot for Petal Length
plt.subplot(2, 2, 3)
sns.boxplot(x='species', y='petal_length', data=dataset, color='green')
plt.title('Petal Length', fontsize=16)
plt.xlabel('Species')
plt.ylabel('Petal Length')
plt.grid(True)

# Subplot boxplot for Petal Width
plt.subplot(2, 2, 4)
sns.boxplot(x='species', y='petal_width', data=dataset, color='orange')
plt.title('Petal Width', fontsize=16)
plt.xlabel('Species')
plt.ylabel('Petal Width')
plt.grid(True)

# Adjusting the plot gaps
plt.subplots_adjust(wspace=0.4, hspace=0.4)

# Showing the main plot
plt.show()
```
